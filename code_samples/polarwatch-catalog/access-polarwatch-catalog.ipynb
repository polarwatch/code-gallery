# Exploring the PolarWatch Catalog

The PolarWatch catalog is organized using the SpatioTemporal Asset Catalog (STAC) standard, a widely used specification for discovering and accessing geospatial data.

Most PolarWatch datasets are available through ERDDAP or HTTPS endpoints, which allow users to access and download data.  These endpoints can also be accessed programmatically, making them suitable for integration into scientific workflows.

In this notebook, we demonstrate how to use the PySTAC Python library to navigate the PolarWatch STAC catalog, explore collections and items, and locate dataset access points such as ERDDAP services or HTTPS download URLs.

For additional examples on how to subset and download data using ERDDAP, please visit the [CoastWatch training site](https://coastwatch-training.github.io/).



## Import Libraries

We begin by importing the PySTAC library.

```python
# !pip install pystac

import pystac  # STAC client library
```

## Load and explore the PolarWatch STAC Catalog

We load the PolarWatch STAC catalog directly from its online location. 


```python
# Load the STAC catalog
catalog = pystac.Catalog.from_file("https://shospital.github.io/stac-vue/stac/catalog.json")

# List Collections in the catalog
collections = list(catalog.get_children())

# Print collection IDs
for c in collections:
    print(c.id)
```

## Accessing Links

The catalog includes useful links for each collection, such as metadata pages and ERDDAP service endpoints.  
The examples below demonstrate how to view and access these links.

```python

coll = catalog.get_child("noaacwSARnrcsnpoleDaily")

# # All links
# for link in coll.links:
#     print(f"{link.rel:20s} {link.href}")

# Specific links 

# about the data
about = next(l for l in coll.links if l.rel == "about")
print(f'About: {about.href}')   

# metadata
describedby = next(l for l in coll.links if l.rel == "describedby")
print(f'Metadata: {describedby.href}' )   

# service
service = next(l for l in coll.links if l.rel == "service")
print(f'Service: {service.href}')  


```

```python
# List collections with ERDDAP service

for coll in catalog.get_children():
    if coll.extra_fields.get("pw:service_type") == "erddap":
        print(coll.id, coll.assets["files"].href)
       
```

```python
# List collections with HTTPS data links with data files
for coll in catalog.get_children():
    if coll.extra_fields.get("pw:service_type") == "https":
        files = coll.assets["files"].href
        print(f"{coll.id}: {files}")
        for item in coll.get_items():
            print(f"  {item.id}: {item.assets['data'].href}")
```
